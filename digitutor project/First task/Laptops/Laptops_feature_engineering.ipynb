# Feature Engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('laptops.csv',index_col=0)
dataset.head()
```

```python
data=dataset.copy()
data['Ram(GB)']=[word.strip('GB') for word in data['Ram']]
data['Ram(GB)']=pd.to_numeric(data['Ram(GB)'])
data['Weight(kg)']=[word.strip('kg') for word in data['Weight']]
data['Weight(kg)']=pd.to_numeric(data['Weight(kg)'])
data.drop(['Weight','Ram'],axis=1,inplace=True)
data.head(50)
```

```python
numerical_features=['Inches','Ram(GB)','Weight(kg)','Price_euros']
```

```python
dataset=data.copy()
```

## Handling Rare Categorical Feature

We will remove categorical variables that are present less than 1% of the observations

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp=dataset.groupby(feature)['Price_euros'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
    
    
```

```python
dataset.head(100)
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['Price_euros'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
dataset.head(10)
```

## Feature scaling

```python
scaling_feature=[feature for feature in dataset.columns if feature not in 'Price_euros']
len(scaling_feature)
```

```python
dataset.head()
```

```python
feature_scale=[feature for feature in dataset.columns if feature not in 'Price_euros']
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
scaler.transform(dataset[feature_scale])
```

```python
dataset.head()
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset['Price_euros'].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```
